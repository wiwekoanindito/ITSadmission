```python
#!pip install pyodbc
#!pip install pandas_profiling
#!pip install pandas

import pandas as pd
import pyodbc
import ydata_profiling
```

```python
sql_conn = pyodbc.connect('DRIVER={SQL Server}; \
                            SERVER=LAPTOP-1NTPD1QL\SQLEXPRESS; \
                            DATABASE=OLTP_PenerimaanMandiri; \
                            Trusted_Connection=yes') 
```

```python
cursor = sql_conn.cursor()
cursor.execute("SELECT table_schema, table_name FROM information_schema.tables WHERE table_type = 'BASE TABLE' AND TABLE_CATALOG='OLTP_PenerimaanMandiri'")
schemas_and_tables = cursor.fetchall()

for schema, table in schemas_and_tables:
    sql_query = f"SELECT * FROM {schema}.{table}"
    try:
        df = pd.read_sql(sql_query, sql_conn)
        report = ydata_profiling.ProfileReport(df)
        report.to_file(f"C:/Users/dito/Downloads/{schema}_{table}_profile_report.html")
    except Exception as e:
        print(f"Error processing table {schema}.{table}: {e}")
```
